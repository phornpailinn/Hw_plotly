```python
import numpy as np
import pandas as pd
```

```python
df = "https://raw.githubusercontent.com/phornpailinn/Hw_plotly/main/Sleep_health_and_lifestyle_dataset.csv"
df =pd.read_csv(df,sep =",")
df
```

```python
df.drop(columns=['Person ID','Sleep Disorder'],axis=1,inplace = True)
```

```python
df['Occupation'].value_counts()
```

```python
QualityofSleep_by_occupation = df.groupby('Occupation', as_index=False)['Quality of Sleep'].mean()
QualityofSleep_by_occupation
```

```python
SleepDuration_by_occupation = df.groupby('Occupation', as_index=False)['Sleep Duration'].mean()
SleepDuration_by_occupation
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

labels = ["Accountant", "Doctor", "Engineer", "Lawyer", "Manager", "Nurse",
          "	Sales Representative","Salesperson","Scientist","Software Engineer","Teacher"]

# Create subplots: use 'domain' type for Pie subplot
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
fig.add_trace(go.Pie(labels=labels, values=[7.891892, 6.647887, 8.412698, 7.893617, 7.000000, 7.369863, 4.000000, 6.000000, 5.000000, 6.500000,6.975000], name="Quality of Sleep"),
              1, 1)
fig.add_trace(go.Pie(labels=labels, values=[7.113514, 6.970423, 7.987302, 7.410638, 6.900000, 7.063014, 5.900000, 6.403125, 6.000000, 6.750000, 6.690000], name="Sleep Duration"),
              1, 2)

# Use `hole` to create a donut-like pie chart
fig.update_traces(hole=.4, hoverinfo="label+percent+name")

fig.update_layout(
    title_text="Quality and Daration of sleep",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Quality', x=0.18, y=0.5, font_size=20, showarrow=False),
                 dict(text='Duration', x=0.82, y=0.5, font_size=20, showarrow=False)])
fig.show()
```
